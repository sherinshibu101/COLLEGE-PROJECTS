```python
def extract_frames(video_path, output_dir):
    os.makedirs(output_dir, exist_ok=True)
    cap = cv2.VideoCapture(video_path)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    interval = max(1, total_frames // FRAMES_PER_VIDEO)

    for i in range(FRAMES_PER_VIDEO):
        cap.set(cv2.CAP_PROP_POS_FRAMES, i * interval)
        ret, frame = cap.read()
        if ret:
            frame_path = os.path.join(output_dir, f"{os.path.basename(video_path).split('.')[0]}_frame_{i:03d}.jpg")
            cv2.imwrite(frame_path, frame)
    cap.release()

def process_dataset(input_dir, output_dir):
    print(f"Processing {input_dir}...")
    for video_file in os.listdir(input_dir):
        video_path = os.path.join(input_dir, video_file)
        if os.path.isfile(video_path):
            extract_frames(video_path, output_dir)

# Call it for real and fake videos
process_dataset(real_videos_dir, real_frames_dir)
process_dataset(fake_videos_dir, fake_frames_dir)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, Subset
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
import os
import random
import numpy as np
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Set random seeds for reproducibility
seed = 42
torch.manual_seed(seed)
np.random.seed(seed)
random.seed(seed)

IMAGE_SIZE = 224

def apply_blending_artifacts(image, **kwargs):
    if random.random() < 0.3:
        h, w = image.shape[:2]
        mask = np.random.rand(h//10, w//10) > 0.8
        mask = cv2.resize(mask.astype(np.uint8), (w, h), interpolation=cv2.INTER_NEAREST)
        noise = np.random.randint(0, 50, image.shape, dtype=np.uint8)
        image = np.where(mask[..., None], noise, image)
    return image

train_transform = A.Compose([
    A.RandomResizedCrop(IMAGE_SIZE, IMAGE_SIZE, scale=(0.2, 1.0)),
    A.HorizontalFlip(p=0.5),
    A.Lambda(name="synthetic_artifacts", image=apply_blending_artifacts),
    A.ColorJitter(brightness=0.3, contrast=0.3, p=0.7),
    A.GaussianBlur(blur_limit=(3, 5), p=0.4),
    A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ToTensorV2()
])

test_transform = A.Compose([
    A.Resize(IMAGE_SIZE, IMAGE_SIZE),
    A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ToTensorV2()
])

class DeepfakeDataset(Dataset):
    def __init__(self, image_paths, labels, transform=None):
        self.image_paths = image_paths
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        label = self.labels[idx]
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        if self.transform:
            img = self.transform(image=img)['image']
        return img, torch.tensor(label, dtype=torch.long)

# Collect all image paths and labels
def get_image_paths_and_labels(real_dir, fake_dir):
    image_paths = []
    labels = []
    for f in os.listdir(real_dir):
        image_paths.append(os.path.join(real_dir, f))
        labels.append(1)
    for f in os.listdir(fake_dir):
        image_paths.append(os.path.join(fake_dir, f))
        labels.append(0)
    return np.array(image_paths), np.array(labels)


#Model: Multi-Scale Local Attention & Cross-Attention Fusion

class MultiScaleLocalAttention(nn.Module):
    def __init__(self, in_channels, scales=[1, 2, 4]):
        super().__init__()
        self.scales = scales
        self.branches = nn.ModuleList([
            nn.Sequential(
                nn.Conv2d(in_channels, 16, kernel_size=3, stride=s, padding=1, dilation=s),
                nn.GELU(),
                nn.BatchNorm2d(16),
                nn.AdaptiveAvgPool2d((14, 14))
            ) for s in scales
        ])
        self.fusion = nn.Conv2d(16 * len(scales), 32, kernel_size=1)

    def forward(self, x):
        outs = [branch(x) for branch in self.branches]
        out = torch.cat(outs, dim=1)
        out = self.fusion(out)
        return out

class CrossAttentionFusion(nn.Module):
    def __init__(self, dim_q, dim_kv, num_heads=4):
        super().__init__()
        self.attn = nn.MultiheadAttention(embed_dim=dim_q, kdim=dim_kv, vdim=dim_kv, num_heads=num_heads, batch_first=True)
        self.norm = nn.LayerNorm(dim_q)

    def forward(self, query, key_value):
        attn_output, _ = self.attn(query, key_value, key_value)
        return self.norm(query + attn_output)

class HybridDeepfakeDetector(nn.Module):
    def __init__(self):
        super().__init__()
        from transformers import ViTModel, ViTConfig
        config = ViTConfig.from_pretrained("WinKawaks/vit-tiny-patch16-224")
        config.output_attentions = True
        self.vit = ViTModel.from_pretrained("WinKawaks/vit-tiny-patch16-224", config=config)
        for name, param in self.vit.named_parameters():
            if 'encoder.layer.9' in name or 'encoder.layer.10' in name or 'encoder.layer.11' in name:
                param.requires_grad = True
            else:
                param.requires_grad = False
        self.local_attn = MultiScaleLocalAttention(3)
        self.cross_attn = CrossAttentionFusion(dim_q=192, dim_kv=32)
        self.classifier = nn.Sequential(
            nn.Linear(192 + 32, 128),
            nn.ReLU(),
            nn.Linear(128, 2)
        )

    def forward(self, x):
        vit_out = self.vit(x)
        global_feat = vit_out.last_hidden_state[:, 0]  # (B, 192)
        local_feat_map = self.local_attn(x)  # (B, 32, 14, 14)
        local_feat = local_feat_map.flatten(2).mean(-1)  # (B, 32)
        global_feat_exp = global_feat.unsqueeze(1)  # (B, 1, 192)
        local_feat_exp = local_feat.unsqueeze(1)    # (B, 1, 32)
        fused_feat = self.cross_attn(global_feat_exp, local_feat_exp).squeeze(1)  # (B, 192)
        features = torch.cat([fused_feat, local_feat], dim=1)
        return self.classifier(features)

def compute_metrics(y_true, y_pred, y_prob, labels=[0,1]):
    report_str = classification_report(y_true, y_pred, target_names=['Fake', 'Real'], digits=3)
    report_dict = classification_report(y_true, y_pred, target_names=['Fake', 'Real'], digits=3, output_dict=True)
    cm = confusion_matrix(y_true, y_pred, labels=labels)
    auc = roc_auc_score(y_true, y_prob[:,1])
    return report_str, report_dict, cm, auc

def plot_confusion_matrix(cm, labels=['Fake', 'Real'], title='Confusion Matrix'):
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
    disp.plot(cmap='Blues')
    plt.title(title)
    plt.show()

def evaluate(model, loader, device):
    model.eval()
    y_true, y_pred, y_prob = [], [], []
    with torch.no_grad():
        for inputs, labels in loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            probs = F.softmax(outputs, dim=1).cpu().numpy()
            preds = outputs.argmax(dim=1).cpu().numpy()
            y_true.extend(labels.cpu().numpy())
            y_pred.extend(preds)
            y_prob.extend(probs)
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    y_prob = np.array(y_prob)
    report_str, report_dict, cm, auc = compute_metrics(y_true, y_pred, y_prob)
    return y_true, y_pred, y_prob, report_str, report_dict, cm, auc

def train_model(model, train_loader, val_loader, device, num_epochs=25):
    optimizer = torch.optim.AdamW(model.parameters(), lr=2e-4)
    criterion = nn.CrossEntropyLoss()
    best_acc = 0.0
    for epoch in range(num_epochs):
        model.train()
        running_loss, correct, total = 0.0, 0, 0
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
        _, _, _, val_report_str, val_report_dict, val_cm, val_auc = evaluate(model, val_loader, device)
        val_acc = val_report_dict['accuracy']
        print(f"Epoch {epoch+1}: Loss {running_loss/len(train_loader):.4f}, Val Acc {val_acc:.4f}, Val AUC {val_auc:.4f}")
        if val_acc > best_acc:
            best_acc = val_acc
            torch.save(model.state_dict(), 'best_model.pth')
    print("Training Complete")

def cross_validate(model_class, image_paths, labels, device, k=5, num_epochs=10, batch_size=16):
    skf = StratifiedKFold(n_splits=k, shuffle=True, random_state=42)
    reports, cms, aucs = [], [], []
    for fold, (train_idx, val_idx) in enumerate(skf.split(image_paths, labels)):
        print(f"\n--- Fold {fold+1}/{k} ---")
        train_dataset = DeepfakeDataset(image_paths[train_idx], labels[train_idx], train_transform)
        val_dataset = DeepfakeDataset(image_paths[val_idx], labels[val_idx], test_transform)
        train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=2, pin_memory=True)
        val_loader = DataLoader(val_dataset, batch_size=batch_size, num_workers=2, pin_memory=True)
        model = model_class().to(device)
        train_model(model, train_loader, val_loader, device, num_epochs=num_epochs)
        model.load_state_dict(torch.load('best_model.pth'))
        _, _, _, report_str, report_dict, cm, auc = evaluate(model, val_loader, device)
        reports.append(report_dict)
        cms.append(cm)
        aucs.append(auc)
        print(f"Fold {fold+1} Accuracy: {report_dict['accuracy']:.4f}, AUC: {auc:.4f}")
        plot_confusion_matrix(cm)
    avg_acc = np.mean([r['accuracy'] for r in reports])
    avg_auc = np.mean(aucs)
    print(f"\nAverage CV Accuracy: {avg_acc:.4f}, Average CV AUC: {avg_auc:.4f}")
    return reports, cms, aucs

if __name__ == "__main__":
    # Set your dataset base path
    base_path = "/content/drive/MyDrive/SDFVD Small-scale Deepfake Forgery Video Dataset/SDFVD"
    real_frames_dir = os.path.join(base_path, "real_frames")
    fake_frames_dir = os.path.join(base_path, "fake_frames")

    # Collect all image paths and labels
    image_paths, labels = get_image_paths_and_labels(real_frames_dir, fake_frames_dir)

    # Split into train_val and test set (e.g., 80% train_val, 20% test)
    train_val_paths, test_paths, train_val_labels, test_labels = train_test_split(
        image_paths, labels, test_size=0.2, stratify=labels, random_state=42
    )

    print(f"Train/Val size: {len(train_val_paths)}, Test size: {len(test_paths)}")

    # Device
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

    # Cross-validation on train/val set only
    reports, cms, aucs = cross_validate(
        HybridDeepfakeDetector,
        train_val_paths,
        train_val_labels,
        device,
        k=5,
        num_epochs=5,
        batch_size=16
    )

    # Final training on all train/val data
    full_train_dataset = DeepfakeDataset(train_val_paths, train_val_labels, train_transform)
    full_train_loader = DataLoader(full_train_dataset, batch_size=16, shuffle=True, num_workers=2, pin_memory=True)
    model = HybridDeepfakeDetector().to(device)
    # Use a validation split from train_val if you want early stopping, or just train for fixed epochs
    train_model(model, full_train_loader, full_train_loader, device, num_epochs=5)
    model.load_state_dict(torch.load('best_model.pth'))

    # Evaluation on held-out test set
    test_dataset = DeepfakeDataset(test_paths, test_labels, test_transform)
    test_loader = DataLoader(test_dataset, batch_size=16, num_workers=2, pin_memory=True)
    test_y_true, test_y_pred, test_y_prob, test_report_str, test_report_dict, test_cm, test_auc = evaluate(model, test_loader, device)
    print("\nTest Set Results:")
    print(f"Accuracy: {test_report_dict['accuracy']:.4f}, ROC-AUC: {test_auc:.4f}")
    print("Classification Report:")
    print(test_report_str)  # Use the pre-generated string report
    plot_confusion_matrix(test_cm)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, Subset
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
import os
import random
import numpy as np
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Set random seeds for reproducibility
seed = 42
torch.manual_seed(seed)
np.random.seed(seed)
random.seed(seed)

IMAGE_SIZE = 224

def apply_blending_artifacts(image, **kwargs):
    if random.random() < 0.3:
        h, w = image.shape[:2]
        mask = np.random.rand(h//10, w//10) > 0.8
        mask = cv2.resize(mask.astype(np.uint8), (w, h), interpolation=cv2.INTER_NEAREST)
        noise = np.random.randint(0, 50, image.shape, dtype=np.uint8)
        image = np.where(mask[..., None], noise, image)
    return image

train_transform = A.Compose([
    A.RandomResizedCrop(IMAGE_SIZE, IMAGE_SIZE, scale=(0.2, 1.0)),
    A.HorizontalFlip(p=0.5),
    A.Lambda(name="synthetic_artifacts", image=apply_blending_artifacts),
    A.ColorJitter(brightness=0.3, contrast=0.3, p=0.7),
    A.GaussianBlur(blur_limit=(3, 5), p=0.4),
    A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ToTensorV2()
])

test_transform = A.Compose([
    A.Resize(IMAGE_SIZE, IMAGE_SIZE),
    A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ToTensorV2()
])

class DeepfakeDataset(Dataset):
    def __init__(self, image_paths, labels, transform=None):
        self.image_paths = image_paths
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        label = self.labels[idx]
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        if self.transform:
            img = self.transform(image=img)['image']
        return img, torch.tensor(label, dtype=torch.long)

# Collect all image paths and labels
def get_image_paths_and_labels(real_dir, fake_dir):
    image_paths = []
    labels = []
    for f in os.listdir(real_dir):
        image_paths.append(os.path.join(real_dir, f))
        labels.append(1)
    for f in os.listdir(fake_dir):
        image_paths.append(os.path.join(fake_dir, f))
        labels.append(0)
    return np.array(image_paths), np.array(labels)


#Model: Multi-Scale Local Attention & Cross-Attention Fusion

class MultiScaleLocalAttention(nn.Module):
    def __init__(self, in_channels, scales=[1, 2, 4]):
        super().__init__()
        self.scales = scales
        self.branches = nn.ModuleList([
            nn.Sequential(
                nn.Conv2d(in_channels, 16, kernel_size=3, stride=s, padding=1, dilation=s),
                nn.GELU(),
                nn.BatchNorm2d(16),
                nn.AdaptiveAvgPool2d((14, 14))
            ) for s in scales
        ])
        self.fusion = nn.Conv2d(16 * len(scales), 32, kernel_size=1)

    def forward(self, x):
        outs = [branch(x) for branch in self.branches]
        out = torch.cat(outs, dim=1)
        out = self.fusion(out)
        return out

class CrossAttentionFusion(nn.Module):
    def __init__(self, dim_q, dim_kv, num_heads=4):
        super().__init__()
        self.attn = nn.MultiheadAttention(embed_dim=dim_q, kdim=dim_kv, vdim=dim_kv, num_heads=num_heads, batch_first=True)
        self.norm = nn.LayerNorm(dim_q)

    def forward(self, query, key_value):
        attn_output, _ = self.attn(query, key_value, key_value)
        return self.norm(query + attn_output)

class HybridDeepfakeDetector(nn.Module):
    def __init__(self):
        super().__init__()
        from transformers import ViTModel, ViTConfig
        config = ViTConfig.from_pretrained("facebook/deit-tiny-distilled-patch16-224")
        config.output_attentions = True
        self.vit = ViTModel.from_pretrained("facebook/deit-tiny-distilled-patch16-224", config=config)
        for name, param in self.vit.named_parameters():
            if 'encoder.layer.9' in name or 'encoder.layer.10' in name or 'encoder.layer.11' in name:
                param.requires_grad = True
            else:
                param.requires_grad = False
        self.local_attn = MultiScaleLocalAttention(3)
        self.cross_attn = CrossAttentionFusion(dim_q=192, dim_kv=32)
        self.classifier = nn.Sequential(
            nn.Linear(192 + 32, 128),
            nn.ReLU(),
            nn.Linear(128, 2)
        )

    def forward(self, x):
        vit_out = self.vit(x)
        global_feat = vit_out.last_hidden_state[:, 0]  # (B, 192)
        local_feat_map = self.local_attn(x)  # (B, 32, 14, 14)
        local_feat = local_feat_map.flatten(2).mean(-1)  # (B, 32)
        global_feat_exp = global_feat.unsqueeze(1)  # (B, 1, 192)
        local_feat_exp = local_feat.unsqueeze(1)    # (B, 1, 32)
        fused_feat = self.cross_attn(global_feat_exp, local_feat_exp).squeeze(1)  # (B, 192)
        features = torch.cat([fused_feat, local_feat], dim=1)
        return self.classifier(features)

def compute_metrics(y_true, y_pred, y_prob, labels=[0,1]):
    report_str = classification_report(y_true, y_pred, target_names=['Fake', 'Real'], digits=3)
    report_dict = classification_report(y_true, y_pred, target_names=['Fake', 'Real'], digits=3, output_dict=True)
    cm = confusion_matrix(y_true, y_pred, labels=labels)
    auc = roc_auc_score(y_true, y_prob[:,1])
    return report_str, report_dict, cm, auc

def plot_confusion_matrix(cm, labels=['Fake', 'Real'], title='Confusion Matrix'):
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
    disp.plot(cmap='Blues')
    plt.title(title)
    plt.show()

def evaluate(model, loader, device):
    model.eval()
    y_true, y_pred, y_prob = [], [], []
    with torch.no_grad():
        for inputs, labels in loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            probs = F.softmax(outputs, dim=1).cpu().numpy()
            preds = outputs.argmax(dim=1).cpu().numpy()
            y_true.extend(labels.cpu().numpy())
            y_pred.extend(preds)
            y_prob.extend(probs)
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    y_prob = np.array(y_prob)
    report_str, report_dict, cm, auc = compute_metrics(y_true, y_pred, y_prob)
    return y_true, y_pred, y_prob, report_str, report_dict, cm, auc

def train_model(model, train_loader, val_loader, device, num_epochs=25):
    optimizer = torch.optim.AdamW(model.parameters(), lr=2e-4)
    criterion = nn.CrossEntropyLoss()
    best_acc = 0.0
    for epoch in range(num_epochs):
        model.train()
        running_loss, correct, total = 0.0, 0, 0
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
        _, _, _, val_report_str, val_report_dict, val_cm, val_auc = evaluate(model, val_loader, device)
        val_acc = val_report_dict['accuracy']
        print(f"Epoch {epoch+1}: Loss {running_loss/len(train_loader):.4f}, Val Acc {val_acc:.4f}, Val AUC {val_auc:.4f}")
        if val_acc > best_acc:
            best_acc = val_acc
            torch.save(model.state_dict(), 'best_model.pth')
    print("Training Complete")

def cross_validate(model_class, image_paths, labels, device, k=5, num_epochs=10, batch_size=16):
    skf = StratifiedKFold(n_splits=k, shuffle=True, random_state=42)
    reports, cms, aucs = [], [], []
    for fold, (train_idx, val_idx) in enumerate(skf.split(image_paths, labels)):
        print(f"\n--- Fold {fold+1}/{k} ---")
        train_dataset = DeepfakeDataset(image_paths[train_idx], labels[train_idx], train_transform)
        val_dataset = DeepfakeDataset(image_paths[val_idx], labels[val_idx], test_transform)
        train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=2, pin_memory=True)
        val_loader = DataLoader(val_dataset, batch_size=batch_size, num_workers=2, pin_memory=True)
        model = model_class().to(device)
        train_model(model, train_loader, val_loader, device, num_epochs=num_epochs)
        model.load_state_dict(torch.load('best_model.pth'))
        _, _, _, report_str, report_dict, cm, auc = evaluate(model, val_loader, device)
        reports.append(report_dict)
        cms.append(cm)
        aucs.append(auc)
        print(f"Fold {fold+1} Accuracy: {report_dict['accuracy']:.4f}, AUC: {auc:.4f}")
        plot_confusion_matrix(cm)
    avg_acc = np.mean([r['accuracy'] for r in reports])
    avg_auc = np.mean(aucs)
    print(f"\nAverage CV Accuracy: {avg_acc:.4f}, Average CV AUC: {avg_auc:.4f}")
    return reports, cms, aucs

if __name__ == "__main__":
    # Set your dataset base path
    base_path = "/content/drive/MyDrive/SDFVD Small-scale Deepfake Forgery Video Dataset/SDFVD"
    real_frames_dir = os.path.join(base_path, "real_frames")
    fake_frames_dir = os.path.join(base_path, "fake_frames")

    # Collect all image paths and labels
    image_paths, labels = get_image_paths_and_labels(real_frames_dir, fake_frames_dir)

    # Split into train_val and test set (e.g., 80% train_val, 20% test)
    train_val_paths, test_paths, train_val_labels, test_labels = train_test_split(
        image_paths, labels, test_size=0.2, stratify=labels, random_state=42
    )

    print(f"Train/Val size: {len(train_val_paths)}, Test size: {len(test_paths)}")

    # Device
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

    # Cross-validation on train/val set only
    reports, cms, aucs = cross_validate(
        HybridDeepfakeDetector,
        train_val_paths,
        train_val_labels,
        device,
        k=5,
        num_epochs=5,
        batch_size=16
    )

    # Final training on all train/val data
    full_train_dataset = DeepfakeDataset(train_val_paths, train_val_labels, train_transform)
    full_train_loader = DataLoader(full_train_dataset, batch_size=16, shuffle=True, num_workers=2, pin_memory=True)
    model = HybridDeepfakeDetector().to(device)
    # Use a validation split from train_val if you want early stopping, or just train for fixed epochs
    train_model(model, full_train_loader, full_train_loader, device, num_epochs=5)
    model.load_state_dict(torch.load('best_model.pth'))

    # Evaluation on held-out test set
    test_dataset = DeepfakeDataset(test_paths, test_labels, test_transform)
    test_loader = DataLoader(test_dataset, batch_size=16, num_workers=2, pin_memory=True)
    test_y_true, test_y_pred, test_y_prob, test_report_str, test_report_dict, test_cm, test_auc = evaluate(model, test_loader, device)
    print("\nTest Set Results:")
    print(f"Accuracy: {test_report_dict['accuracy']:.4f}, ROC-AUC: {test_auc:.4f}")
    print("Classification Report:")
    print(test_report_str)  # Use the pre-generated string report
    plot_confusion_matrix(test_cm)
```
